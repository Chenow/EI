### Arbre de Decision


1. Importer les librairies classiques : numpy, pandas, seaborn et matplolib


```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn
```

2. Lire le fichier iris_data_V2.csv avec pandas


```python
datas = pd.read_csv('iris_data_V2.csv', sep=",")
```

3. Lire les trois premières lignes du fichier. Donner le nombre de lignes et de colonnes


```python
print(datas.shape)
print(datas.values[:3, :])
```

4. Créer cinq listes(array) sepal_length,sepal_widthpetal_length,petal_width et target ou species

```python
sepal_length, sepal_width, petal_length, petal_width, target = datas['sepal_length'], datas['sepal_width'], datas['petal_length'], datas['petal_width'], datas['target']
```

5. Créer une listes(array) contenant sepal_length,sepal_widthpetal_length,petal_width


```python
#tab = [sepal_length, sepal_width, petal_length, petal_width]
tab = datas[['sepal_length','sepal_width' ,'petal_length', 'petal_width']]
```

6. Importer le modèle "tree", from sklearn import tree <


```python
from sklearn import tree
```

6. Creer le modèle < tree.DecisionTreeClassifier()> et faire l'entrainement


```python
from sklearn.model_selection import train_test_split

arbre = tree.DecisionTreeClassifier(criterion="entropy", max_depth=3)
X_train, X_test, y_train, y_test = train_test_split(tab, target, test_size=0.2, random_state=1) 

arbre.fit(X_train, y_train)
tree.plot_tree(arbre)
```

7. Tracer l'arbre < tree.plot_tree() >


```python
from sklearn import metrics
print("Accuracy:",arbre.score(X_test, y_test))
```

8)Verifier sepal_length=4.7,sepal_width=3.2, petal_length=1.3,petal_width=0.2


```python
X_test = [[4.7, 3.2, 1.3, 0.2]]

y_pred = arbre.predict(X_test)
print(y_pred)
```
